### Image processing using OpenCV

- Two samples-> apple and potato (sliced)
- Two images of each of the samples are captured. One by keeping the light source (mobile flash) at approximately 90 degree to the sample and another one by keeping the light source at 53 degree. 

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import skimage.io as sk
```

### Image of slice of apple at 90 degree (day 1)  

```python
apple = cv.imread('apple_1_0.jpg')
resized_apple = cv.resize(apple, (500,500), interpolation = cv.INTER_CUBIC)
apple_rgb = cv.cvtColor(apple, cv.COLOR_BGR2RGB)
sk.imshow(apple_rgb)
```

```python
cropped_apple = resized_apple[150:300,100:300]
cropped_apple_rgb = cv.cvtColor(cropped_apple, cv.COLOR_BGR2RGB)
sk.imshow(cropped_apple_rgb)
```

### Histogram of gray image 

```python
gray_apple = cv.cvtColor(cropped_apple, cv.COLOR_BGR2GRAY)
gray_hist_apple = cv.calcHist([gray_apple],channels = [0],mask = None,histSize = [256],
                        ranges = [0,256])
plt.plot(gray_hist_apple)
plt.xlim([0,256])
plt.show()
```
